```python
#importamos el paquete de pandas asignandole un sobrenombre pd
import pandas as pd

#creamos listas vacias, que seran almacenadas con datos:

#lis1=los últimos 5 días de las ventas
lis1=[]

#lis2=ventas de los dias respectivamente
lis2=[]

#lis3=los valores de las ventas con un descuento del 5%
lis3=[]

#Pedimos al usuario que ingrese el ultimo día de ventas mayor o igual a 5 dias.
n=int(input('Ingrese el último día de ventas mayor o igual a 5: '))

#Ponemos la condición si cumple con lo establecido desarrolla todo el codigo, si no muestra en pantalla que
#'El último día de ventas ingresado es incorrecto.'

#Condición para el valor ingresado de n sea mayor igual que 5.
if n >=5:

#el valor de i recorre desde el dia (n-5,n)=[n-5,n-4,n-3,n-2,n-1]    
    for i in range(n-5,n):
        
#se agregan a la lista lis1 los valores de los dias+1=[n-4,n-3,n-2,n-1,n]
        lis1.append(i+1)
    
#pedimos al usuario que ingrese las ventas de cada dia desde el dia (n-4 a n).   
        v=float(input('Ingrese las ventas del día ' + str(i+1)+':'))
    
#se agregan a las lista lis2 las ventas de cada dia.    
        lis2.append(v)
    
#creamos un diccionario con las tres listas ya almacenadas con datos.        
    dic1={'Día':lis1,
    'Ventas':lis2}

#creamos nuestro dataframe con el diccionario    
    df1=pd.DataFrame(dic1)
    
#imprime el dataframe
    print('Ventas sin descuento')
    print(df1)

#ahora para crear otro dataframe pero con las ventas descontadas en un 5%
#hacemos recorrer a j para todos los valores de la lista lis2
    for j in lis2:
#almacenamos en otra lista lis3 los valores de la lista lis2 multiplicados por 0.95  
        lis3.append(j*0.95)

#creamos otro diccionario con la lista lis3 valores con descuento
    dic2={'Día':lis1,
        'Ventas':lis3}
    
#creamos nuestro dataframe con el nuevo diccionario
    df2=pd.DataFrame(dic2)
    
#imprime el dataframe 
    print('Ventas con descuento')
    print(df2)    
else:
    print('El último día de ventas ingresado es incorrecto.')


```
