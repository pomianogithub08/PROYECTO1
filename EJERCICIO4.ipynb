```python
import pandas as pd
#levantamos las data con pandas
data_riesgos=pd.read_csv('Data - Riesgos.csv')
#genera un dataframe con la 'Data - Riesgos.csv'
df=pd.DataFrame(data_riesgos)

#dimensiones del dataframe
print('Las dimensión del dataframe es: ' , df.shape)

#número de datos que contiene el dataframe
print('El número de datos del dataframe es: ', df.size)

#los nombres de sus columnas
print('Los nombres de sus columnas son:\n ', df.columns)

#los nombres de sus filas
print('Los nombres de sus filas son:\n ', df.index)

#los tipos de datos de las columnas
print('Los tipos de datos de las columnas son:\n ', df.dtypes)

#muestra las 10 primeras filas
print('Las 10 primeras filas son:\n ', df.head(10))

#muestra las 10 últimas filas
print('Las 10 ultimas filas son:\n ', df.tail(10))

#veirificación si hay datos perdidos
a=pd.isnull(df)
print('Verificación de datos perdidos:\n',a)

#total de datos perdidos por columana
b=df.isnull().sum()
print('Total de datos perdidos por columna:\n' ,b)

m=df.iloc[0:100,]
print('Muestra las 100 primeras filas:\n' ,m)
```
